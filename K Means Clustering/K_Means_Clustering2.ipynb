### Let's start a little Easy peasy

```
from sklearn.datasets import make_blobs
```

```
data = make_blobs(n_samples = 250, n_features = 2, centers = 5, cluster_std = 1)
```

```
import matplotlib.pyplot as plt
```

```
plt.scatter(data[0][:, 0], data[0][:, 1])
```

5 clsuters. Directly visble. Right?

```
from sklearn.cluster import KMeans
```

```
model = KMeans(n_clusters = 5)
```

```
model.fit(data[0])
```

```
model.cluster_centers_
```

```
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.set_title('original')
ax1.scatter(data[0][:, 0], data[0][:, 1])
ax2.set_title('KMeans output')
ax2.scatter(data[0][:, 0], data[0][:, 1], c = model.labels_)
```

### Let's make it a little hard for the algorithm

```
data = make_blobs(n_samples = 250, n_features = 2, centers = 5, cluster_std = 2)
```

```
plt.scatter(data[0][:, 0], data[0][:, 1])
```

Can you spot 5 clusters here

```
from sklearn.cluster import KMeans
```

```
model = KMeans(n_clusters = 5)
```

```
model.fit(data[0])
```

```
model.cluster_centers_
```

```
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.set_title('original')
ax1.scatter(data[0][:, 0], data[0][:, 1])
ax2.set_title('KMeans output')
ax2.scatter(data[0][:, 0], data[0][:, 1], c = model.labels_)
```
